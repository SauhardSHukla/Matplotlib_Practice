```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(0, 1, 1000)

# Plot histogram
counts, bin_edges, patches = plt.hist(data, bins=10, edgecolor='black')
# print(counts)
# print(bin_edges)
# print(patches)

# Customizing x-ticks
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])  # Calculate bin centers
plt.xticks(bin_centers, labels=[f'{edge:.2f}' for edge in bin_centers], rotation=45)

# Add labels and title
plt.xlabel('Bin Ranges')
plt.ylabel('Frequency')
plt.title('Histogram with Custom X-Ticks')

# Show plot
plt.show()
```

#### Problem statement
Given File:

amazon_jobs_dataset.csv
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

##### Problem Statement :
Plot the Pie chart between Indian cities vs No. of job openings.

Print the Indian cities and %age of Job distribution in India up to 2 decimal places.

Note: %age of Job distribution should be in descending order.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path  = 'C:/Users/5530/Downloads/amazon_jobs_dataset.csv'
Amazon_jobs = pd.read_csv(file_path)
Amazon_jobs.head(2)
```

```python
Amazon_jobs_loc = Amazon_jobs['location'].str.split(',',expand=True)
Amazon_india = Amazon_jobs_loc[0]=='IN'
city = Amazon_jobs_loc[Amazon_india][2]
city_value = city.value_counts()

#For the plotting of the graph in the pie chart Data
explode=[0.2,0.1,0.1,0.1,0.1]
plt.pie(city_value,autopct='%.2f%%',labels=city_value.index,explode=explode)
plt.axis('equal')
plt.show()
```

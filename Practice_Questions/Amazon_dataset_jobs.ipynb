##### Problem statement
Given File:

amazon_jobs_dataset.csv
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

###### Problem Statement :
Plot the line graph between no. of Job postings with respect to year.

Print the year and the number of job posting as integer value.

Note: Year should be in ascending order.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path  = 'C:/Users/5530/Downloads/amazon_jobs_dataset.csv'
amazon_file = pd.read_csv(file_path)
# amazon_file.head(2)
amazon_file['Year'] = pd.to_datetime(amazon_file['Posting_date']).dt.year
```

```python
value_count = amazon_file['Year'].value_counts()
# for i,J in value_count.items():
#   print(i,J)
sort_index = value_count.sort_index()

plt.plot(sort_index.index,sort_index.values,"b-",marker ='o')
plt.grid()
plt.title('line Graph for the Amazon Job')
plt.xlabel('Year')
plt.ylabel('Jobs Numbers')
plt.show()
```

```python
for i,j in sort_index.items():
  print(f'Year-> {i}:{j}')
```

```python
x=[23,4,3,4]
y=[2,3,4,5]
plt.scatter(x,y)
plt.show()
```

```python
x=[23,4,3,4]
y=[2,3,4,5]
plt.scatter(x,y)
plt.plot(x,y,"b--")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create an array of x values using numpy
x = np.array([0, 1, 2, 3, 4, 5, 6, 7, 9, 10])

# Step 2: Create an array of corresponding y values using numpy
y = np.array([1, 2, 5, 10, 17, 26, 37, 50, 82, 101])

# Step 3: Plot the x and y values
plt.plot(x, y)

# Step 4: Add a grid to the plot for better visualization
plt.grid()

# Step 5: Display the plot
plt.show()

# Step 6: Output the string '60 to 70' (this is separate from the plot logic)
print('60 to 70')
```

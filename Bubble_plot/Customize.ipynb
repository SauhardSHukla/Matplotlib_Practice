Customize the Graph with the help of the matplotlib lib form the python 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Adding the 'color' and 'marker' to the plot

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='')
plt.show()
```

```python
#Adding the linewidth to the graph

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='',linewidth='5')
plt.show()
```

```python
#Adding Various Lable to the graph

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='',linewidth='5')
plt.ylabel('X^2')
plt.xlabel('X Np_Array')
plt.show()
```

```python
#Adding title to the graph

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='',linewidth='5')
plt.ylabel('X^2')
plt.xlabel('X: Np_Array')
plt.title('Practice Graph')
plt.show()
```

```python
#Adding label to the axis

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='',linewidth='5',label='np_array')
plt.legend() # Help to visualise the Label of axis in the Graph "np_array"
plt.ylabel('X^2')
plt.xlabel('X: Np_Array')
plt.title('Practice Graph')
plt.show()
```

```python
#Adding axis [Removing the default values] to the graph

x=np.arange(0,10,0.1)
y=x**2
plt.plot(x,y ,color ='black',marker ='',linewidth='5')
plt.axis([0,10,0,100])
plt.ylabel('X^2')
plt.grid() #To make a grid on the table
plt.text(7,80,'VALUE IS HERE',fontsize =12)  #Add the text to a particalur point in the Graph to markout the particular value or important stuff in the Graph
plt.xlabel('X: Np_Array')
plt.title('Practice Graph')
plt.show()
```

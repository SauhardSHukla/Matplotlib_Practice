Bubble plot  is the MatplotLib in which it work on the scatter plot just using the size to get the dot in the size of given data 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

import numpy as np
import matplotlib.pyplot as plt

year = np.array([2014, 2013, 2012, 2010, 2011])
population = np.array([100, 200, 300, 400, 450])
salary = np.array([20, 30, 3, 34, 45])
c = np.arange(len(year))

# Create scatter plot
plt.scatter(salary, year, s=population, c=c, edgecolors='black', marker='^')

# Set axis ranges
plt.axis([0, 50, 2009, 2015])  # Adjusted to include data points comfortably

# Add grid
plt.grid()

# Add text annotations
for i in c:
    plt.text(salary[i]+0.3, year[i]+0.3, str(year[i]) + ' VALUE TO IT', fontsize=9, ha='center')

# Label axes and add title
plt.ylabel('Year')
plt.xlabel('Salary')
plt.title('Practice Graph')

# Display the graph
plt.show()
```

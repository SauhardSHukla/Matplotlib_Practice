Matplotlib is used to get the data in Visual form that help to handle the various data in the graphical form using pyplot form matplotlib:-

We use the get the data in the form of the pie which hold various data in a visaul form to understand data in better format 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Using various value_measure to make the pie presentable and better to understand

data =[2,3,4,5]
labels = ['A','B','C','D']
color =['blue','green','yellow','red']
explode =[0.1,0.1,0.1,0.2]
plt.pie(data,colors=color,labels=labels,autopct='%.1f%%',explode=explode,counterclock=True,startangle=120,pctdistance=0.7)
plt.axis('equal')
plt.show()
```
